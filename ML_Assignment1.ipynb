```python
import numpy as np
import matplotlib.pyplot as plt

learning_rate = np.load("learning_rate.npy")
train_hist = np.load("train_hist.npy")
val_hist = np.load("val_hist.npy")
epochs = np.arange(1,10)

```

```python
#generate plots
fig, axs = plt.subplots(2,2, figsize=(10,10))
#plot 1
steps = np.arange(learning_rate.shape[0])
axs[0,0].plot(learning_rate, label="learning rate")
axs[0,0].set_xlabel("Training Steps")
axs[0,0].set_ylabel("Learning Rate")
steps = np.arange(learning_rate.shape[0])
#set up epoch dots in plot 1
epoch_len = 391
epoch_dots = np.arange(epoch_len, epoch_len*9+1, epoch_len)
axs[0,0].scatter(epoch_dots, learning_rate.ravel()[epoch_dots-1], marker="o", label="epoch marker")
axs[0,0].legend()
axs[0,0].tick_params(axis="x", rotation=30)
#plot 2
axs[0,1].plot(epochs, train_hist[:,0], "-o", label="train")
axs[0,1].plot(epochs, val_hist[:,0], "-o", label="val")
axs[0,1].set_xticks(epochs)
axs[0,1].set_xlabel("Epochs")
axs[0,1].set_ylabel("Loss")
axs[0,1].legend()
axs[0,1].tick_params(axis="x", rotation=30)
#plot 3
axs[1,0].plot(epochs, train_hist[:,1], "-o", label="train")
axs[1,0].plot(epochs, val_hist[:,1], "-o", label="val")
axs[1,0].set_xticks(epochs)
axs[1,0].set_xlabel("Epochs")
axs[1,0].set_ylabel("Accuracy")
axs[1,0].legend()
axs[1,0].tick_params(axis="x", rotation=30)
#plot 4
axs[1,1].plot(epochs, train_hist[:,2], "-o", label="train")
axs[1,1].plot(epochs, val_hist[:,2], "-o", label="val")
axs[1,1].set_xticks(epochs)
axs[1,1].set_xlabel("Epochs")
axs[1,1].set_ylabel("F1 Score")
axs[1,1].legend()
axs[1,1].tick_params(axis="x", rotation=30)
```
